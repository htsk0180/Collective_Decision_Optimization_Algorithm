```python
# Collective Decision Optimization Algorithm
import random
import numpy as np
import matplotlib.pyplot as plt
import math 

# step: 1 Initialize a population
population = [random.randint(0, 1) for i in range(0, 10)]
print("Başlangıç Popülasyonu: ", population)

# step: 2 termination criterion (T)
T = 100

# step: 3 Calculate the fitness of each search agent
def fitness(population):
    fitness = 0
    for i in range(0, len(population)):
        fitness += population[i]
    return fitness

# The personal best location ← Pop, t=1;
pbest = population
print("Kişisel en iyi konum: ", pbest)

# while t ≤ T do
t = 1
while t <= T:
    # Find the global best (φL)
    gbest = pbest
    print("Global en iyi konum: ", gbest)
    # Compute the step size (stepsize(t)) on stepsize(t) = 2 - 1.7*(t-1 / T-1)
    stepsize = 2 - 1.7*(t-1 / T-1)
    # for i = 1 to N do
    for i in range(0, len(population)):
        new_population = []
        # if φL > φi then
        if fitness(gbest) > fitness(population):
            # Calculate the new solutions (newXq) by the newXq = φL +−→Wq , (q = 1, 2, 3, 4, 5)
            for j in range(0, len(population)):
                new_population.append(gbest[j] + stepsize * random.uniform(-1, 1))
                # newpop ← newXq
                population = new_population
        else:
            # Change the location of an agent by newXi0 = Xi(t) +−→τ0 × stepsize(t) × d0, d0 = φP − Xi(t)
              d0 = fitness(pbest) - population[i]
              new_population.append(population[i] + random.uniform(-1, 1) * stepsize * d0)
            # newpop ← [newXi0, newXi1, newXi2, newXi3, newXpi4]
              population = new_population
        
        # Evaluate the fitness of newpop
        new_fitness = fitness(population)
        # Update Xi(t) and the personal best location using the best one among these resultant positions
        if new_fitness > fitness(pbest):
            pbest = population
    # t=t+1;
    t += 1
# Output the best candidate solution
print("En iyi aday çözüm: ", pbest)

 

    


```
